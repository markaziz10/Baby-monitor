```python
pie = birth.drop(birth.index[[9,10]])
```

```python
pie
```

```python
plt.figure(figsize=(20,12))
# plot chart
ax1 = plt.subplot(121, aspect='equal')
pie.plot(kind='pie', y = '2018', ax=ax1, autopct='%1.1f%%', 
 startangle=90, shadow=False, labels=pie['Area'], legend = False, fontsize=14)

# plot table
ax2 = plt.subplot(122)
plt.axis('off')
ax2.set_title('Area versus 2018 births in the UK')
plt.show()
```

```python
plt.figure(figsize=(20,12))
# plot chart
ax1 = plt.subplot(121, aspect='equal')
pie.plot(kind='pie', y = '2017', ax=ax1, autopct='%1.1f%%', 
 startangle=90, shadow=False, labels=pie['Area'], legend = False, fontsize=14)

# plot table
ax2 = plt.subplot(122)
plt.axis('off')
ax2.set_title('Area versus 2018 births in the UK')
plt.show()
```
